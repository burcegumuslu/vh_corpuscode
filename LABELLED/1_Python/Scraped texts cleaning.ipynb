```python
import json
with open('../scraped/mainstream scraped/deleted_list 20230705', 'r') as file:
    deleted_list = json.load(file)
    
with open('../scraped/mainstream scraped/passed_list 20230705', 'r') as file:
    passed_list = json.load(file)
    
with open('../scraped/mainstream scraped/error_list 20230705', 'r') as file:
    error_list = json.load(file)
    
tried_list = deleted_list + passed_list + error_list
```

```python
import json
import os

directory = "../scraped/mainstream scraped"
for filename in os.listdir(directory):
    if 'json' in filename and filename not in tried_list:
        file_path = os.path.join(directory, filename)
        try: 
            with open(file_path) as f:
                data = json.load(f)

            result_id = data['result_id']
            URL = data['URL']
            language = data['meta']['lang']
            text = data['cleaned_text']

            if len(text) > 500 and language == 'en' and data['URL'] == data['meta']['canonical']:
                with open(f'../scraped/mainstream cleaned/{data["result_id"]}.json', 'w+') as output_file:
                    json.dump(data, output_file)
                passed_list.append(data['result_id'])
            else: 
                deleted_list.append(result_id)
        except:
            print('Error with ', filename)
            error_list.append(filename)

print('Completed')
```

```python
with open('../scraped/mainstream scraped/deleted_list 20230713', 'w+') as file:
    json.dump(deleted_list, file)
    
with open('../scraped/mainstream scraped/passed_list 20230713', 'w+') as file:
    json.dump(passed_list, file)
    
with open('../scraped/mainstream scraped/error_list 20230713', 'w+') as file:
    json.dump(error_list, file)
```
