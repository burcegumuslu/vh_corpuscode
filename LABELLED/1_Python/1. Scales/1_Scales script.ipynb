```python
d = open('camvaccine.txt', 'r') 
##survey 
text = d.read()
```

```python
print(text)

print('complete')
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

tokened_text = word_tokenize(text)
```

```python
print(tokened_text)
```

```python
stop_words = set(stopwords.words("english"))

filtered_text = []

for word in tokened_text:
    if word.lower() not in stop_words:
        filtered_text.append(word.lower())
        
        
#filtered_text = [word.lower() for word in filtered_text if word.isalpha()]
        
print(filtered_text)
```

```python
_1_grams = nltk.ngrams(filtered_text, 1)

_1_grams = [word for word in _1_grams if word[0].isalpha()]

_2_grams = nltk.ngrams(filtered_text, 2)

_2_grams = [word for word in _2_grams if word[0].isalpha() and word[1].isalpha()]

_3_grams = nltk.ngrams(filtered_text, 3)

_3_grams = [word for word in _3_grams if word[0].isalpha() and word[1].isalpha() and word[2].isalpha()]

_4_grams = nltk.ngrams(filtered_text, 4)

_4_grams = [word for word in _4_grams if word[0].isalpha() and word[1].isalpha() and word[2].isalpha() and word[3].isalpha()]
```

```python
from collections import Counter

ngram_counts = Counter(_1_grams)

for ngram, count in ngram_counts.items():
    print(' '.join(ngram), ':', count)

```

```python
ngram_counts = Counter(_2_grams)

for ngram, count in ngram_counts.items():
    print(' '.join(ngram), ':', count)

```

```python
for gram in _3_grams:
    if 'vaccine' in gram:
        print(gram)
```
