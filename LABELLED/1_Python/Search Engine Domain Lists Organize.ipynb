```python
import json

with open('Domain extraction Search Engine/search_results_Bing_CAM_keywords_2023-05-23.json') as d:
    input = json.load(d)
```

```python
print(input)
```

```python
domain_list = []

for key in input:
    for item in key['domains']:
        if not any(element['domain'] == item['domain'] for element in domain_list):
            domain_list.append({'domain': item['domain'], 'source': [[key['key_word'], key['search-engine']]]})
        else:
            for element in domain_list:
                if element['domain'] == item['domain']:
                    element['source'].append([key['key_word'], key['search-engine']])
                                              
print(domain_list )
```

```python
import pandas as pd

df = pd.DataFrame(domain_list)
```
